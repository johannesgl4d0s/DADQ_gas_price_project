```python
url1 = 'https://agsi.gie.eu/api?country='

url2= '&from=2022-01-01&to=2022-12-03&size=300'
for z in range(1,3):
    for i in range(len(list_of_countries)):
        url= str(url1+list_of_countries[i]+url2)
        response = requests.get(url)
        new = pd.DataFrame.from_dict(response.json()['data'])
        df_test = pd.concat([df_test,new],ignore_index=True)
        print(response.json()['data'][0]['name'])
        time.sleep(1)
```
